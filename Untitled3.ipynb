```python
from warnings import filterwarnings
filterwarnings('ignore')
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import GridSearchCV, train_test_split
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (20, 10)  # for graphs styling
plt.style.use('tableau-colorblind10')  # for graph stying
```

```python
main_df = pd.read_csv("Sample-Superstore.csv")
main_df['Order Date'] = pd.to_datetime(main_df['Order Date'])

# Group by Order Date and sum Sales
daily_sales = main_df.groupby('Order Date')['Sales'].sum().reset_index()
daily_sales.set_index('Order Date', inplace=True)
daily_sales.sort_index(inplace=True)
print("Daily Sales Data:")
print(daily_sales.head())
print("\n")
```

```python
daily_sales.describe()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(daily_sales.index, daily_sales['Sales'])
plt.title('Superstore Sales')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.show()
```

```python
def create_features(df, label=None):
    df['date'] = df.index
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.isocalendar().week
    df['is_weekend'] = df['dayofweek'].apply(lambda x: 1 if x >= 5 else 0)

    # --- Lag Features ---
    df['lag_1'] = df['Sales'].shift(1)
    df['lag_7'] = df['Sales'].shift(7)
    df['lag_30'] = df['Sales'].shift(30)
    df['lag_60'] = df['Sales'].shift(60)

    # --- Sales Difference Features ---
    df['sales_diff_1'] = df['Sales'].diff(1)
    df['sales_diff_7'] = df['Sales'].diff(7)
    df['sales_diff_30'] = df['Sales'].diff(30)
    df['sales_diff_60'] = df['Sales'].diff(60)

    # --- Rolling Statistics Features ---
    df['sales_rolling_mean_1'] = df['Sales'].rolling(window=1).mean().shift(1)
    df['sales_rolling_mean_7'] = df['Sales'].rolling(window=7).mean().shift(1)
    df['sales_rolling_mean_30'] = df['Sales'].rolling(window=30).mean().shift(1)
    df['sales_rolling_mean_60'] = df['Sales'].rolling(window=60).mean().shift(1)

    return df

df_with_features = daily_sales.copy()
df_with_features['Year'] = df_with_features.index.year
df_with_features['Week'] = df_with_features.index.isocalendar().week
df_with_features['Day'] = df_with_features.index.day
df_with_features['WeekDay'] = df_with_features.index.dayofweek
df_with_features['Weekend'] = df_with_features['WeekDay'].isin([5, 6]).astype(int)
df_with_features = create_features(df_with_features)

print(df_with_features.head())
```

```python
df_with_features.corr()
```

```python
df = df_with_features[(df_with_features['Sales'] < 10000) & (df_with_features['Sales'] > 20)].copy()
print(df.head())
print(df.dtypes)
print("Size of cleaned data:", df.size)
```

```python
# Split the data into training and testing sets
train_size = int(0.8 * len(df))
train_df = df.iloc[:train_size].copy()
test_df = df.iloc[train_size:].copy()
```

```python
def split(df, label=None):
    # --- Features Selection ---
    X = df[[
        'dayofweek', 'quarter', 'month', 'year', 'dayofyear',
        'dayofmonth', 'weekofyear', 'is_weekend',
        'lag_1',
        'lag_7',
        'lag_30',
        'lag_60',
        'sales_diff_1',
        'sales_diff_7',
        'sales_diff_30',
        'sales_diff_60',
        'sales_rolling_mean_1',
        'sales_rolling_mean_7',
        'sales_rolling_mean_30',
        'sales_rolling_mean_60'
    ]]

    if label:
        y = df[label]
        return X, y
    return X

# Assuming train_df and test_df have a 'Sales' column
X_train, y_train = split(train_df.copy(), label='Sales')
X_test, y_test = split(test_df.copy(), label='Sales')
```

```python
params = {
    'min_child_weight': [4, 5, 6, 7],
    'gamma': [i / 10.0 for i in range(3, 6)],
    'subsample': [i / 10.0 for i in range(6, 11)],
    'colsample_bytree': [i / 10.0 for i in range(6, 11)],
    'max_depth': [2, 3, 4, 5, 6, 7],
}
# Initialize XGB and GridSearch
xgb_reg = xgb.XGBRegressor(objective='reg:squarederror', n_jobs=-1, random_state=42)
grid = GridSearchCV(xgb_reg, params, cv=2) # Reduced cv for faster execution
grid.fit(X_train, y_train)
gridcv_xgb = grid.best_estimator_
r2 = r2_score(y_test, gridcv_xgb.predict(X_test))
print(f"\nR2 score on the test set: {r2}")
```

```python
plt.figure(figsize=(12, 6))
xgb.plot_importance(gridcv_xgb, height=0.8)
plt.title('Feature Importance')
plt.show()
```

```python
test_df['Prediction'] = gridcv_xgb.predict(X_test)
df_all = pd.concat([test_df, train_df], sort=False)

plt.figure(figsize=(15, 5))
plt.plot(df_all['Sales'][-len(test_df['Prediction']):], label='Actual')
plt.plot(test_df['Prediction'], label='Prediction')
plt.title('Prediction vs Actuals')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

def mean_absolute_percentage_error(y_true, y_pred):
    """Calculates MAPE given y_true and y_pred"""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape = mean_absolute_percentage_error(y_true=test_df['Sales'], y_pred=test_df['Prediction'])
print(f"\nMean Absolute Percentage Error (MAPE): {mape:.2f}%")
```

```python
def create_features(df, label=None):
    df['date'] = df.index
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.isocalendar().week
    df['is_weekend'] = df['dayofweek'].apply(lambda x: 1 if x >= 5 else 0)

    # --- Lag Features ---
    df['lag_1'] = df['Sales'].shift(1)
    df['lag_7'] = df['Sales'].shift(7)
    df['lag_30'] = df['Sales'].shift(30)
    df['lag_60'] = df['Sales'].shift(60)

    # --- Sales Difference Features ---
    df['sales_diff_1'] = df['Sales'].diff(1)
    df['sales_diff_7'] = df['Sales'].diff(7)
    df['sales_diff_30'] = df['Sales'].diff(30)
    df['sales_diff_60'] = df['Sales'].diff(60)

    # --- Rolling Statistics Features ---
    df['sales_rolling_mean_1'] = df['Sales'].rolling(window=1).mean().shift(1)
    df['sales_rolling_mean_7'] = df['Sales'].rolling(window=7).mean().shift(1)
    df['sales_rolling_mean_30'] = df['Sales'].rolling(window=30).mean().shift(1)
    df['sales_rolling_mean_60'] = df['Sales'].rolling(window=60).mean().shift(1)

    # --- Features Selection ---
    X = df[[
        'dayofweek', 'quarter', 'month', 'year', 'dayofyear',
        'dayofmonth', 'weekofyear', 'is_weekend',
        'lag_1',
        'lag_7',
        'lag_30',
        'lag_60',
        'sales_diff_1',
        'sales_diff_7',
        'sales_diff_30',
        'sales_diff_60',
        'sales_rolling_mean_1',
        'sales_rolling_mean_7',
        'sales_rolling_mean_30',
        'sales_rolling_mean_60'
    ]]

    if label:
        y = df[label]
        return X, y
    return X
```

```python
# Forecast future dates recursively
future_dates = pd.date_range(start=df_all.index.max(), periods=120, freq='D')
future_df = pd.DataFrame(index=future_dates)
future_df['Sales'] = np.nan

def recursive_forecast(history, n_periods):
    predictions = []
    last_date = history.index[-1]
    current_history = history.copy()
    for i in range(n_periods):
        next_date = last_date + pd.Timedelta(days=1)
        next_df = pd.DataFrame(index=[next_date])
        next_df = create_features(pd.concat([current_history[['Sales']], next_df]), label=None)

        next_X = next_df.loc[[next_date]][X_train.columns]
        next_pred = gridcv_xgb.predict(next_X)[0]
        predictions.append(next_pred)

        temp_pred_df = pd.DataFrame({'Sales': [next_pred]}, index=[next_date])
        current_history = pd.concat([current_history, temp_pred_df])
        last_date = next_date

    return pd.Series(predictions, index=pd.date_range(start=test_df.index.max() + pd.Timedelta(days=1), periods=n_periods, freq='D'))

future_predictions_recursive = recursive_forecast(df_all[['Sales']].copy(), 120) # Pass a copy to avoid modifying original
future_df['xgb_Prediction'] = future_predictions_recursive

plt.figure(figsize=(14, 5))
plt.plot(test_df['Prediction'], label='Test Prediction')
plt.plot(future_df['xgb_Prediction'],color='red', label='Future Prediction')
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
